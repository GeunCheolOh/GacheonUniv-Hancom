### Why Markdown?

**난이도 측면에서 마크다운은 거의 모든 문서 작성 도구 중 가장 쉽습니다.** 복잡한 프로그래밍 언어와 달리, 마크다운은 직관적인 기호만으로 문서를 구조화합니다. 제목은 #으로, 강조는 *로, 목록은 -로 표현하는 식입니다. HTML이나 LaTeX 같은 마크업 언어는 여는 태그와 닫는 태그를 정확히 맞춰야 하지만, 마크다운은 몇 가지 간단한 규칙만 익히면 30분 안에 실무에 활용할 수 있습니다.
  
**IT 업계에서 마크다운은 사실상 표준 문서 포맷입니다.** GitHub의 README 파일, 기술 블로그, API 문서, 프로젝트 위키가 모두 마크다운으로 작성됩니다. Notion, Slack, Discord 같은 협업 도구들도 마크다운 문법을 지원합니다. 개발자가 아니더라도 IT 기업에서 일한다면 마크다운으로 작성된 문서를 매일 접하게 됩니다. 마크다운을 알면 팀 내 커뮤니케이션이 원활해지고, 문서 작성 속도가 워드나 구글 독스보다 훨씬 빨라집니다.
  
**LLM 시대에 마크다운의 가치는 더욱 커졌습니다.** ChatGPT나 Claude 같은 AI는 마크다운으로 구조화된 입력을 월등히 잘 이해합니다. "회의록 작성해줘"보다 "다음 구조로 회의록 작성: # 회의 주제, ## 참석자, ## 논의 사항, ## 액션 아이템"처럼 마크다운 구조로 요청하면 훨씬 정확한 결과를 얻습니다. 또한 LLM의 출력도 대부분 마크다운 형식이므로, 이를 읽고 편집하는 능력은 AI 시대의 필수 문해력이 되었습니다.
마크다운은 배우기 쉽고, 어디서나 쓰이며, AI와의 소통을 향상시킵니다. IT 업계에 발을 들이거나 문서 작업 효율을 높이고 싶다면, 마크다운부터 시작하세요.

## Part 1: 마크다운 기초

마크다운은 간단한 문법으로 문서를 작성할 수 있는 마크업 언어입니다.

### 1.1 제목 (Headers)

제목은 `#` 기호로 만듭니다. `#`의 개수로 레벨을 조절합니다.

```markdown
# 제목 1 (가장 큰 제목)
## 제목 2
### 제목 3
#### 제목 4
##### 제목 5
###### 제목 6 (가장 작은 제목)
```

**결과:**

# 제목 1
## 제목 2
### 제목 3

### 2.1 셀 타입 (Cell Types)

1. **Code 셀**: 파이썬 코드를 실행

**셀 타입 변경:**
- `Esc` + `M`: Markdown으로 변경
- `Esc` + `Y`: Code로 변경

`````python
# 이것은 Code 셀입니다
print("Hello, Jupyter!")
`````

### 2.2 필수 단축키

#### 명령 모드 (Command Mode - 파란색 테두리)

셀을 선택하고 `Esc`를 누르면 명령 모드로 진입합니다.

| 단축키 | 기능 |
|--------|------|
| `A` | 위에 셀 추가 |
| `B` | 아래에 셀 추가 |
| `DD` | 셀 삭제 |
| `Z` | 삭제 취소 |
| `M` | Markdown 셀로 변경 |
| `Y` | Code 셀로 변경 |
| `Shift + Up/Down` | 여러 셀 선택 |
| `Shift + M` | 선택한 셀 병합 |

#### 편집 모드 (Edit Mode - 초록색 테두리)

셀을 더블클릭하거나 `Enter`를 누르면 편집 모드로 진입합니다.

| 단축키 | 기능 |
|--------|------|
| `Ctrl + Enter` | 셀 실행 |
| `Shift + Enter` | 셀 실행 후 다음 셀로 이동 |
| `Alt + Enter` | 셀 실행 후 아래에 새 셀 추가 |
| `Tab` | 자동완성 |
| `Shift + Tab` | 도움말 표시 |
| `Ctrl + /` | 주석 토글 |

# Part 1. matplotlib 기본 규칙

matplotlib은 파이썬의 가장 널리 쓰이는 시각화 라이브러리입니다.
seaborn은 matplotlib을 기반으로 통계 시각화를 쉽게 만들 수 있도록 도와주는 라이브러리입니다.

## Figure와 Axes 계층 구조

matplotlib 그래프는 **Figure**와 **Axes** 두 개의 계층으로 구성됩니다.

- **Figure**: 전체 그래프 캔버스입니다. `plt.figure()` 또는 `plt.subplots()`로 생성합니다.
- **Axes**: Figure 안에 위치한 개별 플롯 영역입니다. 실제 데이터가 그려지는 공간입니다.

```python
fig, ax = plt.subplots()   # Figure 1개, Axes 1개 생성
```

하나의 Figure에 여러 Axes를 배치하면 **서브플롯(subplot)**을 만들 수 있습니다.

````python
# 기본 Figure / Axes 구조 확인
fig, ax = plt.subplots(figsize=(6, 4))

ax.set_title('Basic Figure / Axes Structure')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.text(0.5, 0.5, 'This is an Axes', ha='center', va='center',
        fontsize=14, transform=ax.transAxes)

plt.tight_layout()
plt.show()
````

## 선 그래프와 축 꾸미기

그래프를 명확하게 만들려면 반드시 다음 요소를 포함해야 합니다.

- `set_title()` — 그래프 제목
- `set_xlabel()`, `set_ylabel()` — 축 레이블
- `set_xticks()`, `set_yticks()` — 축 눈금 위치 지정
- `legend()` — 범례 (여러 선을 구분할 때)

````python
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x, y_sin, label='sin(x)')
ax.plot(x, y_cos, label='cos(x)')

ax.set_title('Sine and Cosine Curves')
ax.set_xlabel('x (radians)')
ax.set_ylabel('y')
ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
ax.set_xticklabels(['0', 'π/2', 'π', '3π/2', '2π'])
ax.set_yticks([-1, -0.5, 0, 0.5, 1])
ax.legend()

plt.tight_layout()
plt.show()
````

## marker, color, linestyle 옵션

선 그래프의 시각적 스타일을 다양하게 조절할 수 있습니다.

| 파라미터 | 주요 옵션 |
|----------|-----------|
| `color` | `'red'`, `'#1f77b4'`, `'tab:blue'` 등 |
| `linestyle` | `'-'` (실선), `'--'` (파선), `':'` (점선), `'-.'` (혼합) |
| `marker` | `'o'` (원), `'s'` (사각), `'^'` (삼각), `'+'` (십자) |
| `linewidth` | 선 두께 (기본값 1.5) |
| `alpha` | 투명도 0.0 ~ 1.0 |

````python
x = np.linspace(0, 4 * np.pi, 30)

fig, ax = plt.subplots(figsize=(9, 4))

ax.plot(x, np.sin(x),
        color='tab:blue', linestyle='-', marker='o',
        linewidth=2, markersize=5, label='solid + circle')

ax.plot(x, np.sin(x - 0.5) * 0.8,
        color='tab:red', linestyle='--', marker='^',
        linewidth=1.5, markersize=6, alpha=0.8, label='dashed + triangle')

ax.plot(x, np.sin(x - 1.0) * 0.6,
        color='tab:green', linestyle=':', marker='s',
        linewidth=1.5, markersize=5, alpha=0.7, label='dotted + square')

ax.set_title('marker, color, linestyle Options')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()

plt.tight_layout()
plt.show()
````

## subplots — 여러 그래프 한 번에 배치

`plt.subplots(nrows, ncols)`를 사용하면 하나의 Figure에 여러 Axes를 격자 형태로 배치할 수 있습니다.

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
# axes[행, 열] 형태로 각 Axes에 접근합니다.
```

`fig.suptitle()`은 전체 Figure의 제목을 설정하고,
`plt.tight_layout()`은 서브플롯 간 간격을 자동으로 조정합니다.

````python
x = np.linspace(0, 2 * np.pi, 100)
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 18, 36]

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# 상단 좌: 선 그래프
axes[0, 0].plot(x, np.sin(x), color='tab:blue')
axes[0, 0].set_title('Line: sin(x)')
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('sin(x)')

# 상단 우: cosine 선 그래프
axes[0, 1].plot(x, np.cos(x), color='tab:orange')
axes[0, 1].set_title('Line: cos(x)')
axes[0, 1].set_xlabel('x')
axes[0, 1].set_ylabel('cos(x)')

# 하단 좌: 산점도
np.random.seed(42)
scatter_x = np.random.randn(80)
scatter_y = scatter_x * 0.8 + np.random.randn(80) * 0.5
axes[1, 0].scatter(scatter_x, scatter_y, alpha=0.6, color='tab:green')
axes[1, 0].set_title('Scatter Plot')
axes[1, 0].set_xlabel('x')
axes[1, 0].set_ylabel('y')

# 하단 우: 막대 차트
axes[1, 1].bar(categories, values, color='tab:purple')
axes[1, 1].set_title('Bar Chart')
axes[1, 1].set_xlabel('Category')
axes[1, 1].set_ylabel('Value')

fig.suptitle('2x2 Subplots Example', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
````

## theme, grid, cmap

**theme**: `plt.style.use()`로 전체 그래프의 외관 테마를 설정합니다.
대표적인 스타일로 `'seaborn-v0_8-whitegrid'`, `'ggplot'`, `'dark_background'` 등이 있습니다.

**grid**: `ax.grid()`로 격자선을 추가하면 값을 읽기 쉬워집니다.

**cmap (colormap)**: 수치 값을 색상으로 대응시키는 매핑입니다.
히트맵, 등고선 그래프 등에서 자주 사용합니다.

| cmap 이름 | 용도 |
|-----------|------|
| `'viridis'` | 일반 수치 데이터 (기본 추천) |
| `'coolwarm'` | 양수/음수 대비 (상관계수 등) |
| `'Blues'` | 단방향 강도 표현 |
| `'plasma'` | 고대비 시각화 |

````python
plt.style.use('seaborn-v0_8-whitegrid')

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# grid 예시
x = np.linspace(0, 10, 50)
axes[0].plot(x, np.sin(x), color='tab:blue', linewidth=2)
axes[0].set_title('Grid Example (whitegrid theme)')
axes[0].set_xlabel('x')
axes[0].set_ylabel('sin(x)')
axes[0].grid(True, linestyle='--', alpha=0.7)

# cmap 예시 — 두 가지 colormap 비교
data = np.outer(np.arange(10), np.arange(10)) / 81.0
im = axes[1].imshow(data, cmap='coolwarm', aspect='auto')
axes[1].set_title('cmap: coolwarm')
axes[1].set_xlabel('column index')
axes[1].set_ylabel('row index')
plt.colorbar(im, ax=axes[1])

plt.tight_layout()
plt.show()

plt.style.use('default')
````

# Part 2. 데이터 타입과 시각화 선택 기준

## 범주형 vs 수치형

올바른 그래프를 선택하려면 분석 대상 변수의 **데이터 타입**을 먼저 파악해야 합니다.

**범주형 데이터 (Categorical)**

값이 유한한 집합에 속하며, 크기 비교보다 **종류** 구분이 중요한 데이터입니다.

- **명목형 (Nominal)**: 순서가 없는 범주. 예) 꽃 품종(`species`), 성별(`sex`)
- **서열형 (Ordinal)**: 의미 있는 순서가 있는 범주. 예) 객실 등급(`class`), 승선 항구(`embark_town`)

**수치형 데이터 (Numerical)**

숫자로 표현되며, 크기 비교와 연산이 의미 있는 데이터입니다.

- **이산형 (Discrete)**: 정수 단위로 셀 수 있는 값. 예) 생존자 수(`survived`), 형제자매 수(`sibsp`)
- **연속형 (Continuous)**: 실수 범위에서 임의의 값. 예) 꽃잎 길이(`petal_length`), 나이(`age`)

---

## 시각화 선택 가이드

| 분석 목적 | 변수 타입 | 권장 그래프 |
|----------|-----------|-------------|
| 단일 범주 빈도 | 범주형 | 막대 차트, 파이 차트 |
| 단일 수치 분포 | 수치형 | 히스토그램, KDE, 박스플롯, 바이올린 플롯 |
| 두 수치 변수 관계 | 수치형 × 수치형 | 산점도, 등고선 그래프 |
| 두 범주 변수 관계 | 범주형 × 범주형 | 히트맵 (교차표) |
| 수치 변수 + 범주 구분 | 수치형 + 범주형 | 산점도 (hue), 바이올린 플롯 |

# Part 3. 단일 변량 시각화

단일 변량(univariate) 분석은 **하나의 변수**가 어떤 분포나 빈도를 갖는지 파악합니다.

## 범주형 단일 변량

**막대 차트 (Bar Chart / Count Plot)**

각 범주의 **빈도(count)**를 막대 높이로 표현합니다.
seaborn의 `countplot`은 빈도를 자동으로 계산해 줍니다.

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# countplot — titanic 객실 등급별 인원
sns.countplot(data=titanic, x='class', order=['First', 'Second', 'Third'],
              hue='class', palette='Set2', legend=False, ax=axes[0])
axes[0].set_title('Passenger Count by Class (Titanic)')
axes[0].set_xlabel('Passenger Class')
axes[0].set_ylabel('Count')

# 생존율 막대 차트 — 집계 후 bar
survival_rate = titanic.groupby('class')['survived'].mean().reset_index()
survival_rate.columns = ['class', 'survival_rate']
class_order = ['First', 'Second', 'Third']
survival_rate['class'] = pd.Categorical(survival_rate['class'], categories=class_order, ordered=True)
survival_rate = survival_rate.sort_values('class')

axes[1].bar(survival_rate['class'], survival_rate['survival_rate'],
            color=['#2ecc71', '#3498db', '#e74c3c'])
axes[1].set_title('Survival Rate by Class (Titanic)')
axes[1].set_xlabel('Passenger Class')
axes[1].set_ylabel('Survival Rate')
axes[1].set_ylim(0, 1)

plt.tight_layout()
plt.show()
````

**파이 차트 (Pie Chart)**

전체 대비 각 범주의 **비율**을 원형으로 표현합니다.
범주 수가 적고 비율 차이가 명확할 때 효과적입니다.

````python
embark_counts = titanic['embark_town'].value_counts().dropna()

fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(embark_counts.values,
       labels=embark_counts.index,
       autopct='%1.1f%%',
       startangle=90,
       colors=['#3498db', '#2ecc71', '#e74c3c'])
ax.set_title('Titanic Passengers by Embarkation Town')

plt.tight_layout()
plt.show()
````

## 수치형 단일 변량

**히스토그램 (Histogram)**

연속 데이터를 **구간(bin)**으로 나누어 각 구간의 빈도를 막대로 표현합니다.
분포의 형태(좌우 치우침, 봉우리 수)를 파악할 때 사용합니다.

````python
fig, axes = plt.subplots(1, 2, figsize=(11, 4))

axes[0].hist(iris['sepal_length'], bins=20, color='steelblue', edgecolor='white')
axes[0].set_title('Histogram: Sepal Length (bins=20)')
axes[0].set_xlabel('Sepal Length (cm)')
axes[0].set_ylabel('Frequency')

axes[1].hist(iris['sepal_length'], bins=8, color='coral', edgecolor='white')
axes[1].set_title('Histogram: Sepal Length (bins=8)')
axes[1].set_xlabel('Sepal Length (cm)')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
````

**KDE (Kernel Density Estimate)**

히스토그램을 매끄럽게 연속화한 **밀도 곡선**입니다.
bin 크기의 영향을 받지 않아 분포 형태를 더 안정적으로 보여줍니다.
`fill=True`를 사용하면 곡선 아래 영역이 채워집니다.

````python
fig, ax = plt.subplots(figsize=(8, 4))

sns.kdeplot(data=iris, x='petal_length', hue='species',
            fill=True, alpha=0.4, ax=ax)
ax.set_title('KDE: Petal Length by Species (Iris)')
ax.set_xlabel('Petal Length (cm)')
ax.set_ylabel('Density')

plt.tight_layout()
plt.show()
````

**박스플롯 (Box Plot)**

**중앙값, 사분위수, 이상값**을 한 번에 표현합니다.
여러 변수나 집단의 분포를 비교할 때 유용합니다.

- 상자의 중앙선: 중앙값 (Q2)
- 상자의 위아래 경계: Q3 (75%) / Q1 (25%)
- 수염(whisker): Q1 − 1.5×IQR ~ Q3 + 1.5×IQR
- 점: 이상값(outlier)

````python
numeric_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 전체 수치 컬럼 비교
sns.boxplot(data=iris[numeric_cols], palette='Set3', ax=axes[0])
axes[0].set_title('Box Plot: Iris Numeric Columns')
axes[0].set_xlabel('Feature')
axes[0].set_ylabel('Value (cm)')

# species별 petal_length 비교
sns.boxplot(data=iris, x='species', y='petal_length',
            hue='species', palette='Set2', legend=False, ax=axes[1])
axes[1].set_title('Box Plot: Petal Length by Species')
axes[1].set_xlabel('Species')
axes[1].set_ylabel('Petal Length (cm)')

plt.tight_layout()
plt.show()
````

**바이올린 플롯 (Violin Plot)**

박스플롯과 KDE를 결합한 형태입니다.
박스플롯이 제공하는 사분위수 정보에 더해 **분포의 형태**를 함께 보여줍니다.
데이터가 이봉 분포(두 개의 봉우리)를 갖는 경우 바이올린 플롯에서 이를 확인할 수 있습니다.

````python
fig, ax = plt.subplots(figsize=(8, 5))

sns.violinplot(data=iris, x='species', y='petal_length',
               hue='species', palette='Set2', legend=False, inner='box', ax=ax)
ax.set_title('Violin Plot: Petal Length by Species')
ax.set_xlabel('Species')
ax.set_ylabel('Petal Length (cm)')

plt.tight_layout()
plt.show()
````

# Part 4. 다중 변량 시각화

다중 변량(multivariate) 분석은 **두 개 이상의 변수** 사이의 관계를 파악합니다.

## 히트맵 (Heatmap)

두 변수의 교차 정보를 **색상 강도**로 표현합니다.
상관계수 행렬에 가장 자주 사용하며, `annot=True`로 수치 값을 셀 내부에 표시할 수 있습니다.

- 1에 가까울수록 강한 양의 선형 관계
- −1에 가까울수록 강한 음의 선형 관계
- 0에 가까울수록 선형 관계 없음

````python
corr = iris.select_dtypes(include='number').corr()

fig, ax = plt.subplots(figsize=(7, 5))
sns.heatmap(corr, annot=True, fmt='.2f',
            cmap='coolwarm', vmin=-1, vmax=1,
            linewidths=0.5, ax=ax)
ax.set_title('Correlation Heatmap: Iris Features')

plt.tight_layout()
plt.show()
````

## 산점도 (Scatter Plot)

두 수치형 변수의 관계를 **점**으로 표현합니다.
seaborn의 `scatterplot`에서 `hue` 파라미터를 사용하면 범주형 변수에 따라 점 색상을 구분해
**3개 변수를 동시에** 시각화할 수 있습니다.

````python
fig, ax = plt.subplots(figsize=(8, 5))

sns.scatterplot(data=iris, x='sepal_length', y='petal_length',
                hue='species', palette='Set1',
                alpha=0.8, s=60, ax=ax)
ax.set_title('Scatter Plot: Sepal Length vs Petal Length by Species')
ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Petal Length (cm)')
ax.legend(title='Species')

plt.tight_layout()
plt.show()
````

## 등고선 그래프 (Contour Plot)

두 수치형 변수의 **결합 밀도(joint density)**를 등고선 형태로 표현합니다.
색상이 진할수록 해당 구간에 데이터가 밀집해 있음을 의미합니다.

`plt.contourf`는 채워진(filled) 등고선을 그리며, `np.meshgrid`로 격자 좌표를 만들어 사용합니다.

````python
x_data = iris['sepal_length'].values
y_data = iris['petal_length'].values

xmin, xmax = x_data.min() - 0.5, x_data.max() + 0.5
ymin, ymax = y_data.min() - 0.5, y_data.max() + 0.5

xx, yy = np.meshgrid(np.linspace(xmin, xmax, 100),
                     np.linspace(ymin, ymax, 100))
positions = np.vstack([xx.ravel(), yy.ravel()])
kernel = stats.gaussian_kde(np.vstack([x_data, y_data]))
zz = kernel(positions).reshape(xx.shape)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 등고선 그래프
cf = axes[0].contourf(xx, yy, zz, levels=12, cmap='Blues')
axes[0].scatter(x_data, y_data, alpha=0.3, s=15, color='navy')
plt.colorbar(cf, ax=axes[0], label='Density')
axes[0].set_title('Contour Plot: Sepal vs Petal Length Density')
axes[0].set_xlabel('Sepal Length (cm)')
axes[0].set_ylabel('Petal Length (cm)')

# 비교용 산점도
species_colors = {'setosa': '#e74c3c', 'versicolor': '#2ecc71', 'virginica': '#3498db'}
for sp, color in species_colors.items():
    mask = iris['species'] == sp
    axes[1].scatter(iris.loc[mask, 'sepal_length'],
                    iris.loc[mask, 'petal_length'],
                    color=color, alpha=0.7, s=40, label=sp)
axes[1].set_title('Scatter Plot: Sepal vs Petal Length by Species')
axes[1].set_xlabel('Sepal Length (cm)')
axes[1].set_ylabel('Petal Length (cm)')
axes[1].legend(title='Species')

plt.tight_layout()
plt.show()
````
